```python
import random


def menu():
    print('''Menu: 
(1) Criar Lista
(2) Ler Lista
(3) Soma da Lista
(4) Media da Lista
(5) Procurar o Maior elemento da Lista
(6) Procurar o Menor elemento da Lista
(7) Ver se a Lista está ordenada por ordem crescente
(8) Ver se a Lista está ordenada por ordem decrescente
(9) Procurar um elemento da Lista
(0) Sair''')
    return

def criarLista(tam):
    listaAlt = []
    i = 0
    while i < tam:
        num = random.randint(1,100)
        listaAlt.append(num)
        i = i + 1
    return listaAlt

def lerLista(tam):
    listaCriada = []
    i = 0
    while i < tam:
        num = int(input("Introduza o elemento " + str(i+1) + " da lista: "))
        listaCriada.append(num)
        i = i + 1 
    return listaCriada

def somaLista(lista):
    soma = 0
    for num in lista:
        soma = soma + num
    return soma

def mediaLista(lista):
    media = somaLista(lista) / len(lista)
    return media

def maiorelemLista(lista):
    maior = lista[0]
    for i in lista:
        if i >= maior:
            maior = i
    return maior

def menorelemLista(lista):
    menor = lista[0]
    for i in lista:
        if i <= menor:
            menor = i
    return menor

def ordCresLista(lista):
    ant = lista[0]
    for i in lista:
        if ant <= i :
            ant = i
        else: return False
    return True

def ordDecresLista(lista):
    ant = lista[0]
    for i in lista:
        if ant >= i:
            ant = i
        else: return False
    return True


menu()
opção = int(input("Selecione a opção: "))
lista = []

while opção != 0:

    if opção ==1:
        lista.clear()
        lista = criarLista(int(input("Qual o tamanho da lista? ")))
        print ("A sua lista é: ", lista )
        opção = int(input("Selecione a opção: "))

    if opção == 2:
        lista.clear()
        lista = lerLista(int(input("Qual o tamanho da lista? ")))
        print ("A sua lista é: ", lista )
        opção = int(input("Selecione a opção: "))
    
    if opção == 3:
        if lista == []:
            print("A soma dos elementos da Lista é: 0")
        else: print("A soma dos elementos da Lista é: ",somaLista(lista))
        opção = int(input("Selecione a opção: "))

    if opção == 4:
        if lista == []:
            print("A média dos elementos da Lista é: 0")
        else: print ("A média dos elementos da Lista é:" , mediaLista(lista))
        opção = int(input("Selecione a opção: "))
    
    if opção == 5:
        if lista == []:
            print("Não existem elementos na Lista")
        else: print("O maior elemento da Lista é: " ,maiorelemLista(lista))
        opção = int(input("Selecione a opção: "))
    
    if opção == 6:
        if lista == []:
            print("Não existem elementos na Lista")
        else: print("O menor elemnto da Lista é: ", menorelemLista(lista))
        opção = int(input("Selecione a opção: "))

    if opção == 7:
        if lista == []:
            print("Não existem elementos na Lista")
        elif ordCresLista(lista) == True:
            print("Sim")
        else: print("Não")
        opção = int(input("Selecione a opção: "))

    if opção == 8:
        if lista == []:
            print("Não existem elementos na Lista")
        elif ordDecresLista(lista) == True:
            print("Sim")
        else: print("Não")
        opção = int(input("Selecione a opção: "))
    
    if opção == 9:
        if lista == []:
            print("Não existem elementos na Lista")
        else: 
            escolha = int(input("Qual o elemento que pretende procurar? "))
            posiçao = lista.index(escolha)
            print("O elemento escolhido encontra-se na posição ", posiçao+1)
        opção = int(input("Selecione a opção: "))


```
